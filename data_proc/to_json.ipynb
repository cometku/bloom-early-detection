```python
lake = 'Erie'
def to_json(df, lake):
    #ONE_FORECAST
    one_forecast = []
    for row in df[['ds', 'chlor_'+lake]].tail(31).itertuples(index = False):
        date, chlor = row[0], row[1]
        day_dict = {
            'x': date,
            'y': chlor
        }
        one_forecast.append(day_dict)
    #ONE_MONTH
    one_month = []
    for row in df[['ds', 'chlor_'+lake]].iloc[555:586].itertuples(index = False):
        date, chlor = row[0], row[1]
        day_dict = {
            'x': date,
            'y': chlor
        }
        one_month.append(day_dict)
    chloroAForecast = df['chlor_'+lake].tail(1).to_numpy()[0]
    return {
        "type": "Feature",
        "properties": {
            "name": lake,
            "ONE_MONTH": one_month,
            "ONE_FORECAST": one_forecast,
            "chloroAForecast": chloroAForecast
        }
    }
```

```python
to_json(df, 'Superior')
```

```python
lakes = ['Erie', 'Superior', 'Huron', 'Ontario', 'Michigan']
```

```python
all_data = []
for lake in lakes:
    all_data.append(to_json(df, lake))
```
